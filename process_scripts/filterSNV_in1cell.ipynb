```python
fi=open('regular.snv')
all_sites=[]
for line in fi:
    seq=line.rstrip().split('\t')
    site=seq[0]+':'+seq[1]+':'+seq[2]
    all_sites.append(site)
fi.close()
#合并的snv深度超过阈值后，还是分开写以记录readInfo
fi=open('regular.snv')
uniq_site={}
for line in fi:
    seq=line.split('\t')
    site=seq[0]+':'+seq[1]+':'+seq[2]
    other_info=':'.join(seq[4:-1])+':'+seq[-1]
    try:
        uniq_site[site].append(other_info)
    except Exception,e:
        uniq_site[site]=[other_info]
fi.close()

results=[]
counter = Counter(all_sites)
for snv,count in counter.items():
    chr_site=snv.split(':')[0]
    site_0=snv.split(':')[1]
    site_1=snv.split(':')[2]
    if count > 10:
        for value in uniq_site[snv]:
            supporting_read, strand, read_info=value.rstrip().split(':')
            results.append("{}\t{}\t{}\t{}\t{}\t{}\t{}".format(chr_site, site_0, site_1, type, supporting_read, strand, read_info)) 

sorted_results = sorted(results, key=lambda x: (x.split()[0], int(x.split()[1]), int(x.split()[2])))
with open('regular.snv_filterDepth_readInfo', 'w') as fo:
    fo.writelines([result + '\n' for result in sorted_results])
```
