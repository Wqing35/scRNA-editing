```r
df_10X<-read.table("/disk1/wenqing/tmp_data/pbmc/result/10X_Ver3/CTSS_aver_depth_100.txt",header=F,sep="\t")
df_s2<-read.table("/disk1/wenqing/tmp_data/pbmc/result/s2/CTSS_aver_depth_100.txt",header=F,sep="\t")
```

```r
head(df_s2)
```

```r
df_10X$group<-"10X"
df_10X$x<-1:dim(df_10X)[1]
colnames(df_10X)<-c("Chr","y","gc","group","x")

df_s2$group<-"Smart-seq2"
df_s2$x<-1:dim(df_s2)[1]
colnames(df_s2)<-c("Chr","y","gc","group","x")
```

```r
head(df_10X)
```

```r
df <- rbind(df_10X,df_s2)
dim(df)
```

```r
library(ggplot2)
```

```r
range(df_10X$y)
range(df_s2$y)
```

```r
p1 <- ggplot(df,aes(x=x,y=y))+
geom_point(aes(color=group))+
geom_line(aes(color=group))+
scale_color_manual(values = c("blue","red")) +  # 自定义颜色
facet_wrap(~group,ncol=1,nrow=2,
scales = "free")+
theme_bw()+
theme(legend.position = "none")+
ylab('Average Depth')+
xlab('scaled length of CTSS')

```

```r

p2 <- ggplot(df,aes(x=x,y=y))+
geom_point(aes(color=group))+
geom_line(aes(color=group))+
scale_color_manual(values = c("blue","red")) +  # 自定义颜色
facet_wrap(~group,ncol=1,nrow=2,
scales = "free")+
theme_bw()+
theme(legend.position = "none")+
ylim(c(0,3))+
ylab('Average Depth')+
xlab('scaled length of CTSS')
```
